1. High-Resolution Optical Imagery

These are used for looking at the ground, measuring vegetation, and tracking urban growth.
| Dataset Name | Snippet ID | Resolution | Best For |
| :--- | :--- | :--- | :--- |
| Sentinel-2 | COPERNICUS/S2_SR_HARMONIZED | 10m | Vegetation health, water quality, and land cover. |
| Landsat 8/9 | LANDSAT/LC08/C02/T1_L2 | 30m | 50+ year historical records and land change analysis. |
| NAIP | USDA/NAIP/DOQQ | 0.6m - 1m | Ultra-high res aerial photos of the US (farm-level detail). |

2. Terrain and Elevation (Topography)

Used to calculate slope, aspect, or visualize mountain ranges.
| Dataset Name | Snippet ID | Resolution | Description |
| :--- | :--- | :--- | :--- |
| SRTM Digital Elevation | USGS/SRTMGL1_003 | 30m | The gold standard for global elevation data. |
| ALOS World 3D | JAXA/ALOS/AW3D30/V3_2 | 30m | Global surface model from JAXA. |

3. Climate and Weather

Global datasets for tracking rainfall, temperature, and atmospheric gas.
| Dataset Name | Snippet ID | Resolution | Best For |
| :--- | :--- | :--- | :--- |
| ERA5-Land | ECMWF/ERA5_LAND/HOURLY | 9km | Precise hourly weather (temp, wind, pressure). |
| CHIRPS Rainfall | UCSB-CHG/CHIRPS/DAILY | 5.5km | Tracking droughts and precipitation over time. |
| Sentinel-5P | COPERNICUS/S5P/OFFL/L3_NO2 | 1.1km | Air quality (Nitrogen Dioxide, Ozone, Methane). |

For example, to load the latest global elevation data and center the map at a particular region:


# Eg : 1) to load the latest global elevation data and 2) center the map at a particular region:





    // Load the global elevation dataset
    var srtm = ee.Image('USGS/SRTMGL1_003');

    // Center the map on the Grand Canyon and display it

    Map.setCenter(-112.11, 36.05, 10);

      -112.11 (Longitude): The East/West coordinate. Negative values are West (this points to the Grand Canyon, USA).

          36.05 (Latitude): The North/South coordinate.

          10 (Zoom Level): This is a scale from 0 to 24.

              0: The entire world.

              10: City level (shows the Grand Canyon and surrounding plateaus).

              20+: Individual buildings or trees.

      2. Map.addLayer(srtm, {min: 0, max: 3000}, 'Elevation');

      This function takes your data (the srtm image) and "paints" it onto the map.
      The Three Parameters:

          srtm (The Object): This is the variable name of the data you want to display.

          {min: 0, max: 3000} (Visualization Parameters): This is the most important part. Raw satellite data isn't an "image" yet—it's just a grid of numbers.

              By default, GEE maps 0 to black and 255 to white.

              Since mountains can be 3,000 meters high, a default view would just look like a solid white blob.

              This code tells GEE: "Make 0 meters look black and 3,000 meters look white." Everything in between will be a shade of gray.

          'Elevation' (The Label): This is just a name. It will appear in the "Layers" tab in the top-right corner of the map, allowing you to toggle the data on and off.



PYTHON IMPLEMENTATION

```python
import ee
import geemap

# 1. Initialize the Earth Engine library
# ee.Authenticate() # Uncomment if running for the first time
ee.Initialize()

# 2. Load the global elevation dataset
srtm = ee.Image('USGS/SRTMGL1_003')

# 3. Create an interactive map
Map = geemap.Map()

# 4. Set the center (Longitude, Latitude, Zoom)
Map.setCenter(-112.11, 36.05, 10)

# 5. Add the layer with visualization parameters
vis_params = {
    'min': 0,
    'max': 3000
}
Map.addLayer(srtm, vis_params, 'Elevation')

# 6. Display the map
Map
```
